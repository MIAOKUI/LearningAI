## 1. 使用kera处理手写数字识别数据集

## 导入相关包

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
from keras.callbacks import TensorBoard
%cd /Users/MIAOKUI/Documents/机器学习/LearningAI/MNIST
```

### 1.2 初始化各种参数


```python
batch_size = 128
num_classes = 10
epochs = 5
img_rows, img_cols = 28,28
```

### 1.3 预处理数据

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data("/Users/MIAOKUI/Documents/机器学习/LearningAI/MNIST/data/mnist.npz")

if K.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)


print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(input_shape)
```

```python
## 数据类型转换
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

## 数据标准化
x_train  = x_train/255
x_test = x_test/255

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

## 把y标签值转换为one hot形式。
y_train = keras.utils.to_categorical(y_train , num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

### 1.4 建立模型
Conv2D(3,3)x32 --> Conv2D(3,3)x64 --> MaxPooling(3,3) --> Dropout(0.25) --> Dense(128) --> Dropout(0.5) --> Dense(10)




```python
model = Sequential()
model.add(Conv2D(32, 
                 kernel_size=(3,3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64,
                 kernel_size=(3,3),
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```


### 1.5 模型训练和模型预测

```python

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test),
          callbacks=[TensorBoard(log_dir='/Users/MIAOKUI/Documents/机器学习/LearningAI/MNIST/tf_log/minst')])
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

### 2. 迁移学习
训练只识别0-6的模型， 固定一部分财顺，再对7-9的数据进行训练

#### 2.1 初始化参数

```python
import datetime
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend

now = datetime.datetime.now
batch_size = 128
num_classes = 5
epochs = 5
img_rows, img_cols = 28, 28
filters = 32
pool_size =2
kernel_size = 3

if K.image_data_format() == 'channels_first':
    input_shape = (1, img_rows, img_cols)
else:
    input_shape = (img_rows, img_cols, 1)
```

#### 2.2 数据预处理

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data("/Users/MIAOKUI/Documents/机器学习/LearningAI/MNIST/data/mnist.npz")

## 创建两个数据集
x_train_lt5 = x_train[y_train <5]
y_train_lt5 = y_train[y_train <5]
x_test_lt5 = x_test[y_test <5]
y_test_lt5 = y_test[y_test <5]

x_train_gte5 = x_train[y_train >= 5]
y_train_gte5 = y_train[y_train >= 5] - 5
x_test_gte5 = x_test[y_test>=5]
y_test_gte5 = y_test[y_test>=5] - 5
```

#### 2.3 创建两组网络，特征网络与分类网络

```python
## 特征提取网络， 用第一个数据集训练完后将不再更新权值
feature_layers = [
    Conv2D(filters, kernel_size, padding='valid', input_shape=input_shape, activation='relu'),
    Conv2D(filters, kernel_size, activation='relu'),
    MaxPooling2D(pool_size=pool_size),
    Dropout(0.25),
    Flatten()
]

## 分类网络
classification_layers = [
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes),
    Activation('softmax')
]

## 创建网络
model = Sequential(feature_layers + classification_layers)
```

#### 2.4 训练网络

```python
def train_model(model, train, test, num_classes):
    x_train = train[0].reshape((train[0].shape[0],) + input_shape)
    x_test = test[0].reshape((test[0].shape[0],) + input_shape)
    x_train = x_train.astype('float32')
    x_test = x_test.astype('float32')
    x_train = x_train/255
    x_test = x_test/255
    print('x_train shape:', x_train.shape)
    print(x_train.shape[0], 'train samples')
    print(x_test.shape[0], 'test samples')
    ## one 表示标签值
    y_train = keras.utils.to_categorical(train[1], num_classes)
    y_test = keras.utils.to_categorical(test[1], num_classes)
    model.compile(loss = 'categorical_crossentropy',
                  optimizer = 'adadelta',
                  metrics = ['accuracy'])
    t = now()
    model.fit(x_train, y_train,
              batch_size = batch_size, 
              epochs = epochs, 
              verbose = 1,
              validation_data = (x_test, y_test))
    print('Training time: %s' %(now() - t))
    score = model.evaluate(x_test, y_test, verbose = 0)
    print(score)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])

train_model(model, 
            (x_train_lt5, y_train_lt5),
            (x_test_lt5, y_test_lt5),
            num_classes)
```

#### 2.5 迁移学习
把feature部分的网络全部冻结，用另一部分数据只训练后面几个分类网络。

```python
## 冻结feature部分的网络。
for l in feature_layers:
    l.trainable = False

## 迁移学习后半部分的数据
train_model(model,
            (x_train_gte5, y_train_gte5), 
            (x_test_gte5, y_test_gte5), 
            num_classes)
```

训练时间可以看到缩短了一半。 结果0.98，还说得过去。
